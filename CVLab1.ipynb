```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
img = mpimg.imread("/content/drive/MyDrive/images.jpg")
```

```python
M,N,C = img.shape
Pixel = M*N
print(Pixel)
```

```python
#to display the image
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.figure()

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1,3,2)
plt.imshow(img_rgb)
plt.title("RGB Image")

plt.subplot(1,3,3)
plt.imshow(img_gray,cmap="gray")
plt.title("Gray Image")
```
